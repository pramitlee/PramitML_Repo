```python
df.drop(['id'],axis=1,inplace=True)
df.head()
```

```python
diag_target=[]
for i in df['diagnosis']:
    if(i=='M'):
        diag_target.append(1);
    elif(i=='B'):
        diag_target.append(0);
        
diag_target
```

```python
df['diag_target']=diag_target
df.head()
```

```python
df.info()
```

```python
df.drop(['diagnosis'],axis=1,inplace=True)
df.head()
```

```python
df.describe()
```

```python
df.hist(figsize=(20,20))
plt.show()
```

```python
X=df.iloc[:,:-1]
X.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
df['diag_target'].value_counts()
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
clf=SVC(kernel='linear')
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
print('Accuracy:{} '.format(accuracy_score(y_test,y_pred)))
print('Classification Report:\n {}'.format(classification_report(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix
print('Confusion Matrix: \n {}'.format(confusion_matrix(y_test,y_pred)))
```

```python
from sklearn.metrics import roc_auc_score
print('ROC-AUC Score: {}'.format(roc_auc_score(y_test,y_pred)))
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
```

```python
print('Precision Score: {}'.format(precision_score(y_test,y_pred)))
print('Recall Score: {}'.format(recall_score(y_test,y_pred)))
```

```python
from sklearn.metrics import log_loss
print('Log Loss: {}'.format(log_loss(y_test,y_pred)))
```

```python
from sklearn.metrics import f1_score
print('F1 Score: {}'.format(f1_score(y_test,y_pred)))
```

### It is important to consider FN rather than FP in this problem. So we are considering β=2

```python
from sklearn.metrics import fbeta_score
print('F2 Score: {}'.format(fbeta_score(y_test,y_pred,beta=2)))
```
