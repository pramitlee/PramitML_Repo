```python
df=pd.read_csv("breast-cancer.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(['id'],axis=1,inplace=True)
df.head()
```

```python
diag_target=[]
for i in df['diagnosis']:
    if(i=='M'):
        diag_target.append(1);
    elif(i=='B'):
        diag_target.append(0);
        
diag_target
```

```python
df['diag_target']=diag_target
df.head()
```

```python
df.info()
```

```python
df.drop(['diagnosis'],axis=1,inplace=True)
df.head()
```

```python
df.describe()
```

```python
df.hist(figsize=(20,20))
plt.show()
```

```python
X=df.iloc[:,:-1]
X.head()
```

```python
y=df.iloc[:,-1]
y.head()
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
```

```python
y_pred=nb.predict(X_test)
y_pred
```

```python
print('Accuracy: {}'.format(accuracy_score(y_test,y_pred)))
print('\n')
print('Classification Report:\n {}'.format(classification_report(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix
print('Confusion Matrix: \n {}'.format(confusion_matrix(y_test,y_pred)))
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
print('Precision Score: {}'.format(precision_score(y_test,y_pred)))
print('Recall Score: {}'.format(recall_score(y_test,y_pred)))
```

```python
from sklearn.metrics import roc_auc_score
print('ROC-AUC Score: {}'.format(roc_auc_score(y_test,y_pred)))
```

```python
from sklearn.metrics import log_loss
print('Log Loss: {}'.format(log_loss(y_test,y_pred)))
```

```python
from sklearn.metrics import f1_score
print('F1 Score: {}'.format(f1_score(y_test,y_pred)))
```

### It is important to consider FN rather than FP in this problem. So we are considering Î²=2

```python
from sklearn.metrics import fbeta_score
print('F2 Score: {}'.format(fbeta_score(y_test,y_pred,beta=2)))
```

```python
from sklearn.metrics import roc_curve 
fpr, tpr, thresholds = roc_curve(y_test, nb.predict_proba(X_test)[:,1]) 
plt.figure(figsize = (12,8)) 
plt.plot(fpr, tpr, label='Naive Bayes (area = %0.2f)' % (roc_auc_score(y_test,y_pred)))
plt.plot([0, 1], [0, 1],'r--') 
plt.text(0.5, 0.5, "varying threshold scores (0-1)", rotation=0, size=12,ha="center", va="center",bbox=dict(boxstyle="rarrow"))
plt.xlabel('False Positive Rate') 
plt.ylabel('True Positive Rate') 
plt.title('Receiver operating characteristic') 
plt.legend(loc="lower right") 
plt.savefig('Naive_Bayes_ROC') 
plt.show()
```
